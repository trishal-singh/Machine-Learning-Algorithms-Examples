```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
%matplotlib inline
#importing data
df =pd.read_csv('cars.csv')
for column in df.columns:
    if df[column].dtype == type(object):
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
#creting model    
y = KMeans(n_clusters=3,max_iter=3000)
kmeans= y.fit_predict(df)
```

```python
# viuali
df = df.as_matrix(columns=None)
plt.scatter(df[kmeans == 0, 0], df[kmeans == 0,1],s=100,c='red',label='US')
plt.scatter(df[kmeans == 1, 0], df[kmeans == 1,1],s=100,c='blue',label='Japan')
plt.scatter(df[kmeans == 2, 0], df[kmeans == 2,1],s=100,c='green',label='Europe')
plt.scatter(y.cluster_centers_[:,0],y.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of car brands')
plt.legend()
plt.show()
```
